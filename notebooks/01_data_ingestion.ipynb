```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path("../data/raw")

csv_files = list(data_path.glob("*.csv"))

print(csv_files)
```

```python
datasets = {}

for file in csv_files:
    df = pd.read_csv(file)

    datasets[file.stem] = df

    print("\n" + "="*50)
    print("Dataset:", file.stem)
    print("Shape:", df.shape)
    print(df.head())
```

```python
for name, df in datasets.items():

    print("\n" + "="*50)
    print("Dataset:", name)

    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())
```
