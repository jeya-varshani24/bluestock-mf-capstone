```python
import pandas as pd
import numpy as np
```

```python
nav = pd.read_csv(
    "../data/raw/02_nav_history.csv"
)

print("Shape:", nav.shape)

nav.head()
```

```python
print("Missing Values:")
print(nav.isnull().sum())

print("\nDuplicates:")
print(nav.duplicated().sum())

print("\nNAV <= 0:")
print((nav['nav'] <= 0).sum())
```

```python
nav['date'] = pd.to_datetime(
    nav['date']
)

print(nav['date'].dtype)
```

```python
nav = nav.sort_values(
    by=['amfi_code', 'date']
)

nav.head()
```

```python
nav['nav'] = (
    nav.groupby('amfi_code')['nav']
    .ffill()
)

print(
    nav['nav'].isnull().sum()
)
```

```python
nav.to_csv(
    "../data/processed/clean_nav.csv",
    index=False
)

print(
    "clean_nav.csv saved successfully!"
)

print(
    "Final Shape:",
    nav.shape
)
```

```python
transactions = pd.read_csv(
    "../data/raw/08_investor_transactions.csv"
)

print("Shape:", transactions.shape)

transactions.head()
```

```python
print("Missing Values:")
print(
    transactions.isnull().sum()
)

print("\nDuplicates:")
print(
    transactions.duplicated().sum()
)

print("\nTransaction Types:")
print(
    transactions['transaction_type']
    .value_counts()
)

print("\nKYC Status:")
print(
    transactions['kyc_status']
    .value_counts()
)

print("\nTransactions with amount <= 0:")
print(
    (transactions['amount_inr'] <= 0)
    .sum()
)
```

```python
transactions['transaction_type'] = (
    transactions['transaction_type']
    .str.strip()
    .str.title()
)

print(
    transactions['transaction_type']
    .value_counts()
)
```

```python
print(
    transactions.columns.tolist()
)
```

```python
transactions['transaction_type'] = (
    transactions['transaction_type']
    .str.strip()
)

transactions['transaction_type'] = (
    transactions['transaction_type']
    .replace({
        'SIP': 'SIP',
        'Sip': 'SIP',
        'Lumpsum': 'Lumpsum',
        'Redemption': 'Redemption'
    })
)

print(
    transactions['transaction_type']
    .value_counts()
)
```

```python
transactions['transaction_date'] = pd.to_datetime(
    transactions['transaction_date']
)

print(
    transactions['transaction_date'].dtype
)
```

```python
valid_kyc = [
    'Verified',
    'Pending'
]

invalid_kyc = transactions[
    ~transactions['kyc_status']
    .isin(valid_kyc)
]

print(
    "Invalid KYC Records:",
    len(invalid_kyc)
)
```

```python
transactions.to_csv(
    "../data/processed/clean_transactions.csv",
    index=False
)

print(
    "clean_transactions.csv saved successfully!"
)

print(
    "Final Shape:",
    transactions.shape
)
```

```python
performance = pd.read_csv(
    "../data/raw/07_scheme_performance.csv"
)

print("Shape:", performance.shape)

performance.head()
```

```python
print(
    performance.columns.tolist()
)
```

```python
numeric_columns = [
    'return_1yr_pct',
    'return_3yr_pct',
    'return_5yr_pct',
    'benchmark_3yr_pct',
    'alpha',
    'beta',
    'sharpe_ratio',
    'sortino_ratio',
    'std_dev_ann_pct',
    'max_drawdown_pct',
    'aum_crore',
    'expense_ratio_pct'
]

performance[numeric_columns] = (
    performance[numeric_columns]
    .apply(pd.to_numeric, errors='coerce')
)

print("Missing values after numeric conversion:")
print(
    performance[numeric_columns]
    .isnull()
    .sum()
)
```

```python
negative_sharpe = performance[
    performance['sharpe_ratio'] < 0
]

print(
    "Funds with Negative Sharpe Ratio:",
    len(negative_sharpe)
)

if len(negative_sharpe) > 0:
    print(
        negative_sharpe[
            ['scheme_name', 'sharpe_ratio']
        ]
    )
```

```python
invalid_expense = performance[
    (performance['expense_ratio_pct'] < 0.1)
    |
    (performance['expense_ratio_pct'] > 2.5)
]

print(
    "Funds with Invalid Expense Ratios:",
    len(invalid_expense)
)

if len(invalid_expense) > 0:
    print(
        invalid_expense[
            ['scheme_name',
             'expense_ratio_pct']
        ]
    )
```

```python
performance.to_csv(
    "../data/processed/clean_performance.csv",
    index=False
)

print(
    "clean_performance.csv saved successfully!"
)

print(
    "Final Shape:",
    performance.shape
)
```
